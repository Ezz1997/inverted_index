```python
import re,os
import numpy as np
import math
import operator
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
#returns a list of casefolded stopwords
def stopwordlist():
    path = os.getcwd()
    fo=open(path+"\stop_words_english.txt","r",encoding='utf-8')
    x=fo.read()
    stoplist=x.split(x.casefold())
    fo.close()
    return stoplist
```

```python
stoplistt = stopwordlist()         #list of casefolded stopwords
```

```python
#creating inverted index
def creatinvertedindex():
    dict={}
    path = os.getcwd()
    files=os.listdir(path+"\information14\\txt")
    qq=re.compile('\W+')
    global filelist
    filelist=[]
    stoplistt=stopwordlist()

    for f in files:
        fname=path+"\information14\\txt\\"
        fname+=f
        lines=open(fname,"r",encoding='utf-8')
        wordslist=lines.read()
        x=qq.split(wordslist.casefold())
        name=f
        for i in x:
            if i in stoplistt:
                continue
            if not i.isalpha():
                continue
            i=porter.stem(i)
            if i not in dict:
                dict[i]=[]
                dict[i].append(name)
            else:
                if name not in dict[i]:
                    dict[i].append(name)
        lines.close()
        filelist.append(name)
        fname=""
    
    wordfile=open("invertedindex.txt","w",encoding='utf-8')
    wordfile.write(str(dict))
    wordfile.close()
    return dict
```

```python
invertndex=creatinvertedindex()
```

```python
#returns a list of words in sorted order so that we can check the index of a particular word

def listofwords():
    lst=list(invertndex.keys())
    lst.sort()
    return lst
```

```python
lstofword = listofwords()    #list of sorted words in entire dataset
```

```python
#returns a list of document frequency of terms
def get_df(): 
    templist=[]
    for i in range(len(lstofword)):
        templist.append(0)
    for i in lstofword:
        print(len(invertndex[i]))
        templist[lstofword.index(i)]=(len(invertndex[i]))
    return templist
```

```python
dflist=get_df()         #list of doc frequency of terms
```

```python
def VSMofDoc():
    dict={}                      #dictionary for VSM
    temp=[]                     #temporary list for document vector
    path = os.getcwd()          #returns the path of the current folder 
    files=os.listdir(path+"\information14\\txt")         #returns all the files present in the specified directory
    qq=re.compile('\W+')                #for splitting words
    global filelist                        # list of file name
    filelist=[]
    max_freq={}
    for f in files:                     # for each file present in the directory
        for jj in range(len(lstofword)):    # initilizing the vector with zero 
            temp.append(0)
        fname=path+"\information14\\txt\\"
        fname+=f                            #file name
        lines=open(fname,"r",encoding='utf-8')               #open a file 
        wordslist=lines.read()              #read the data from the file
        x=qq.split(wordslist.casefold())        # casefolding and then splitting the words
        name=f
        for i in x:
            if i in stoplistt:               #to no to include the stopword in our VSM
                continue
            if not i.isalpha():
                continue
            i=porter.stem(i)                 #stem the word
            indexofterm=lstofword.index(i)      #getting the index of the term
            temp[indexofterm]+=1            #incremening the term frequency
        dict[f]=temp.copy()                 #assigning the vector to its doc id
        max_freq[f]=max(temp)               # max freq of any word in that doc 
        lines.close()                       #closing the file
        filelist.append(name)               #append the name of the file which is processed
        fname=""
        temp.clear()
    
    for ii,j in dict.items():
        for iii in range(len(j)):
            if j[iii]!=0:
                product=( j[iii] /max_freq[ii] )*( math.log( len(filelist)/dflist[iii] ,10 ) )           #mutiplying the term frequency by its idf
                j[iii]=product
    
    #write the VSM to a file  
    wordfile=open("VSM.txt","w",encoding='utf-8')
    wordfile.write(str(dict))
    wordfile.close()
    return dict
    
```

```python
VSM_dict=VSMofDoc()
```

```python
#returns the ranking of the document
import warnings
def calculateranking(query):
    qq=re.compile('\W+')
    weightlist=[]
    query=qq.split(query.casefold())
    parsedquery=[]
    queryvector=[]
    for i in query:
        if i in stoplistt:           #to not to consider the stopword
            continue
        i=porter.stem(i.casefold())       #casefolding and stemming the word of the query
        if i in lstofword:
            parsedquery.append(i)       #return a list of words present both in a query and our term dictionary
        else:
            continue 
        
    for jj in range(len(lstofword)):    # initilizing the vector with zero 
        queryvector.append(0)  
             
    for x in parsedquery:
        queryvector[lstofword.index(x)]+=1      #increment the term frequency 
    
    for ii in range(len(queryvector)):
        idf=( math.log( 56/dflist[ii] ,10) )        #calculate the idf of the paricular term and document
        queryvector[ii]=queryvector[ii]*idf         #multiplying the tf with idf
        
        
    for key,docvector in VSM_dict.items():
        numerator=np.dot(queryvector,docvector)         #calculating the dot product of the query and document
        magnitude_of_Query_vector=np.linalg.norm(queryvector)       #calculating the magnitude of the query vector
        magnitude_of_Doc_vector=np.linalg.norm(docvector)           #calculating the magnitude of the document vector
        denominator=(magnitude_of_Query_vector)*(magnitude_of_Doc_vector)       #ultiplying the mag of query vector and mag of doc vector
        # ans=math.cos(numerator/denominator)
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')   
            ans=numerator/denominator   
        answer=(key,ans)        #doc id and the value of the its similarity
        weightlist.append(answer)
        del answer
        
    weightlist.sort(key=operator.itemgetter(1),reverse=True)        #sorts the weightlist
    templist=[]
    alpha=0.0005
    for y in weightlist:
        if y[1]<alpha:      #filtering out the result by alpha
            break
        # t=(y[0],y[1])
        # templist.append(t)
        templist.append(y[0])
    return templist
        
```

```python
# language model 1 using unigram and no smoothing
query=input("Enter Query: ")

#total = sum(docsLength.values())
#print(total)
stop_words = set(stopwords.words('english'))

allDocsP = {}
prob = 1.0
for key in docsLength.keys():
    for sent in query.split():
        if sent in stop_words:
            continue
        if not sent.isalpha():
            continue
        w = sent.lower()
        w=porter.stem(w)
        if tf[key].get(w) == None:
            prob = prob * (float(0.0)) / docsLength.get(key)
        else:
            prob = prob * (float(tf[key].get(w)) / docsLength.get(key))
                     
    allDocsP[key] = prob
    prob = 1.0

sort_files = sorted(allDocsP.items(), key=lambda x: x[1], reverse=True)

for i in sort_files:
	print(i[0], i[1])  
```

```python
# language model 2 using unigram and Laplace Smoothing with a value of 1
query=input("Enter Query: ")

stop_words = set(stopwords.words('english'))

allDocsP = {}
prob = 1.0
count = 0
for key in docsLength.keys():
    for sent in query.split():
        if sent in stop_words:
            continue
        if not sent.isalpha():
            continue
        w = sent.lower()
        w=porter.stem(w)
        if tf[key].get(w) == None:
            count += 1
            prob = prob * (float(1.0)) / (docsLength.get(key) + count)
        else:
            prob = prob * (float(tf[key].get(w) + 1) / (2 * docsLength.get(key) + count))
                     
    allDocsP[key] = prob
    prob = 1.0

sort_files = sorted(allDocsP.items(), key=lambda x: x[1], reverse=True)

for i in sort_files:
	print(i[0], i[1])   
```
